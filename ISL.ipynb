```python
from tensorflow.keras.applications import InceptionV3

inception = InceptionV3(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
#don't train existing weights
for layer in inception.layers:
    layer.trainable = False
```

```python
#useful for getting number of output classes
folders = glob('D:\data\Digits\*')
```

```python
folders
```

```python
#our layers - you can add more if you want
x = Flatten()(inception.output)
```

```python
prediction = Dense(len(folders), activation='softmax')(x)

#create a model object
model = Model(inputs=inception.input, outputs=prediction)
```

```python
model.summary()
```

```python
#tell the model what cost and optimization method to use
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
#Use the Image Data Generator to import the images from the dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
#Make sure you provide the same target size as initialied for the image size
training_set = train_datagen.flow_from_directory('D:\data\Digits',
                                                 target_size = (128, 128),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
test_set = test_datagen.flow_from_directory('D:\data\Digits',
                                            target_size = (128, 128),
                                            batch_size = 32,
                                            class_mode = 'categorical')
```

```python
#fit the model
r = model.fit(
    training_set,
    epochs=10,
    steps_per_epoch=len(training_set),
    validation_data=test_set,
    validation_steps=len(test_set)
    )
```

```python
import os
import json
from tensorflow.keras.optimizers import Adam

# Next steps: visualize training, save artifacts, evaluate, and (optional) fine-tune.
import matplotlib.pyplot as plt

# 1) Plot training history
def plot_history(history, out_dir='.'):
    h = history.history
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(h.get('accuracy', []), label='train_acc')
    plt.plot(h.get('val_accuracy', []), label='val_acc')
    plt.legend(); plt.title('Accuracy')
    plt.subplot(1,2,2)
    plt.plot(h.get('loss', []), label='train_loss')
    plt.plot(h.get('val_loss', []), label='val_loss')
    plt.legend(); plt.title('Loss')
    os.makedirs(out_dir, exist_ok=True)
    out_path = os.path.join(out_dir, 'training_history.png')
    plt.savefig(out_path)
    plt.close()
    print(f"Saved training plot to: {out_path}")

# 2) Save model and label mapping
def save_model_and_labels(model, train_iterator, out_dir='.'):
    os.makedirs(out_dir, exist_ok=True)
    model_path = os.path.join(out_dir, 'digit_inception_model.h5')
    model.save(model_path)
    # save class indices (label -> folder index)
    class_indices = train_iterator.class_indices
    labels_map = {v:k for k,v in class_indices.items()}  # index -> label
    labels_path = os.path.join(out_dir, 'labels_map.json')
    with open(labels_path, 'w') as f:
        json.dump(labels_map, f, indent=2)
    print(f"Saved model to: {model_path}")
    print(f"Saved labels map to: {labels_path}")

# 3) Evaluate on test set
def evaluate_model(model, test_iterator):
    loss, acc = model.evaluate(test_iterator, steps=len(test_iterator))
    print(f"Test loss: {loss:.4f}  Test accuracy: {acc:.4f}")
    return loss, acc

# 4) Optional: fine-tune some top layers of the base model (InceptionV3)
def fine_tune(model, base_model, train_iterator, val_iterator, fine_layers=50, epochs=5, lr=1e-5):
    # unfreeze the top `fine_layers` layers of the base model
    for layer in base_model.layers[-fine_layers:]:
        layer.trainable = True
    # recompile with a low learning rate
    model.compile(optimizer=Adam(learning_rate=lr),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    history_fine = model.fit(
        train_iterator,
        epochs=epochs,
        steps_per_epoch=len(train_iterator),
        validation_data=val_iterator,
        validation_steps=len(val_iterator)
    )
    return history_fine

# Example usage (adjust out_dir as desired). These variables are already present in the notebook:
# r, model, training_set, test_set, inception, Data_path
out_dir = os.path.join(Data_path, 'model_artifacts')
plot_history(r, out_dir=out_dir)
save_model_and_labels(model, training_set, out_dir=out_dir)
evaluate_model(model, test_set)

# To fine-tune (optional): uncomment and run
# hf = fine_tune(model, inception, training_set, test_set, fine_layers=50, epochs=5, lr=1e-5)
# plot_history(hf, out_dir=out_dir)
# save_model_and_labels(model, training_set, out_dir=out_dir)
```

```python
# inference helpers (uses existing `model`, `training_set`, `test_set`, and `IMAGE_SIZE`)
def _build_labels_map():
    return {v: k for k, v in training_set.class_indices.items()}

def predict_array(arr, model=model, labels_map=None):
    # arr: HxWxC (0-1 float) or 1xHxWxC
    if labels_map is None:
        labels_map = _build_labels_map()
    if arr.ndim == 3:
        arr = np.expand_dims(arr, 0)
    preds = model.predict(arr)
    idx = int(np.argmax(preds, axis=1)[0])
    return labels_map[idx], float(preds[0, idx])

def predict_image(path, model=model, labels_map=None):
    # load_img and image.img_to_array are available in the notebook
    img = load_img(path, target_size=IMAGE_SIZE + [3])
    arr = image.img_to_array(img) / 255.0
    return predict_array(arr, model=model, labels_map=labels_map)

# Example usages:
# 1) Predict on an image file:
# label, prob = predict_image(r'D:\\data\\Digits\\1\\some_image.jpg')
# print(label, prob)

# 2) Predict on the first image from a test batch:
imgs, lbls = next(test_set)
pred_label, pred_prob = predict_array(imgs[0])
print("Predicted:", pred_label, "prob:", pred_prob)
```

```python
import os
import numpy as np

# Quick test cell: evaluate model, inspect a batch of predictions, and run a single-file prediction example.
# Uses existing variables/functions: model, r, training_set, test_set, imgs, lbls, folders, out_dir,
# plot_history, save_model_and_labels, evaluate_model, predict_image, _build_labels_map


# Summary of what we've done (for reference):
# - built InceptionV3 base (weights='imagenet', include_top=False)
# - froze base layers, added Flatten + Dense(num_classes, softmax)
# - compiled with categorical_crossentropy + adam
# - prepared ImageDataGenerator and created training_set/test_set
# - trained the model (history stored in `r`)
# - helper functions for plotting/saving/predicting exist in the notebook

# 1) Evaluate on the full test set
print("=== Full test set evaluation ===")
evaluate_model(model, test_set)

# 2) Inspect predictions on the currently loaded batch `imgs` / `lbls`
labels_map = _build_labels_map()  # index -> label (string)
preds = model.predict(imgs)       # batch predictions
pred_idx = np.argmax(preds, axis=1)
true_idx = np.argmax(lbls, axis=1)

n_show = min(8, imgs.shape[0])
print(f"\n=== Sample predictions (first {n_show} of current batch) ===")
for i in range(n_show):
    print(f"sample {i}: true={labels_map[int(true_idx[i])]}  pred={labels_map[int(pred_idx[i])]}  prob={preds[i, pred_idx[i]]:.4f}")

batch_acc = float(np.mean(pred_idx == true_idx))
print(f"\nBatch accuracy on this loaded batch: {batch_acc:.4f}")

# 3) Example single-image prediction from the dataset (first image in first folder)
print("\n=== Single-image example prediction ===")
try:
    candidate_files = [f for f in os.listdir(folders[0]) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    if candidate_files:
        example_path = os.path.join(folders[0], candidate_files[0])
        label, prob = predict_image(example_path)
        print("Example file:", example_path)
        print("Predicted label:", label, "probability:", prob)
    else:
        print("No image files found in folder:", folders[0])
except Exception as e:
    print("Could not run single-image example:", str(e))

# 4) Save artifacts and plot training history (will overwrite if already present)
print("\n=== Save artifacts and plot training history ===")
plot_history(r, out_dir=out_dir)
save_model_and_labels(model, training_set, out_dir=out_dir)
print("Artifacts saved to:", out_dir)
```
